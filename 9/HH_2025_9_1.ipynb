#2.Meta-Learner 활용

- Treatment Effect(처리효과)
 - ATE (Average Treatment Effect): 전체 집단에 대한 평균 처리 효과. 때때로 중요한 정보를 숨길 수 있음.
 - ATT (Average Treatment effect on the Treated): 실제 처리를 받은 대상 집단에 대한 평균 효과.
 - ATC (Average Treatment effect on the Control): 대조군에 대한 평균 처리 효과.
 - CATE (Conditional Average Treatment Effect)
 - HTE (HTE: Heterogeneous Treatment Effect): 조건부 평균 처리 효과. 단위 간 이질적 반응을 탐구 가능. 서로 다른 특성을 가진 단위가 동일한 처리에 다르게 반응할 수 있음.

<메타러너 기반 모델들>
>S-Learner
- 단일 base learner 사용. 간단하지만 효과적인 모델
- CATE 추정 가능하며 HTE 모델링에 적합
- ATE만으로는 포착되지 않는 정보를 파악 가능
- 가정
 1. 모든 교란 요인이 제어 가능해야 함
 2. 노이즈에 민감
 3. 단순성으로 인해 처리를 무시할 수 있음 → 인과관계 과소평가
 4. 처리 변수를 다른 일반 변수처럼 취급 → 영향 작으면 생략될 수 있음

>T-Learner
- S-Learner의 한계(처리 변수 무시)를 극복하기 위해 고안
- 각 처리 상태별로 별도의 모델을 학습
- 처리 효과가 작을 경우, S-Learner보다 적합
- 특징
 - 처리 변수를 명시적으로 반영
 - 관측치 수가 불균형할 경우 분산 증가 → 특히 데이터 적을 때 취약
 - 데이터가 많아질수록 S-Learner와의 성능 차이 줄어듦

>X-Learner
- T-Learner의 성능 저하를 개선, CATE를 효과적으로 추정, 데이터셋 불균형 상황에서 활용
- 예: 새로운 기능을 일부 사용자에게만 시험 적용할 때
- 특징
 - 샘플이 큰 경우 적합하며, 불균형 데이터일수록 유리
 - 데이터셋이 작을 경우, 추가 모델 피팅 → 노이즈 증가 → 성능 저하 우려

```python
import pandas as pd
import numpy as np
df = pd.read_csv("uplift_train.csv", index_col=[0])
clients = pd.read_csv("clients.csv", index_col=[0])
clients = clients[["age", "gender"]]
```

```python
df1 = df.merge(clients, on='client_id')
df1.gender.value_counts()
```

```python
df1.gender.replace({'U':0, 'F':1, 'M':2}, inplace=True)
df1.head()
```

```python
confounding_col = ['age', 'gender']
```

```python
from sklearn.model_selection import train_test_split
y = "target"
T = "treatment_flg"
X = confounding_col

X_train, X_test, y_train, y_test, trmt_train, trmt_test, train, test = train_test_split( df1[X], df1[y], df1[T], df1)
```

- Regressor로만 계산

```python
from xgboost import XGBRegressor
import numpy as np

# Treatment 별로 모델 분리 (T-Learner)
treated_model = XGBRegressor()
control_model = XGBRegressor()

# 학습
treated_model.fit(train.query("treatment_flg==1")[X], train.query("treatment_flg==1")[y])
control_model.fit(train.query("treatment_flg==0")[X], train.query("treatment_flg==0")[y])

# 예측 (확률값)
treated_pred = treated_model.predict(X_test)  # P(Y=1|T=1, X)
control_pred = control_model.predict(X_test)  # P(Y=1|T=0, X)

# CATE = 두 예측값 차이
cate = treated_pred - control_pred
print(f"Estimated ATE: {np.mean(cate) * 100:.2f}%")
```

- Classifier로 계산

```python
from xgboost import XGBClassifier

# Treatment 별로 모델 분리
treated_model = XGBClassifier()
control_model = XGBClassifier()

# 학습
treated_model.fit(train.query("treatment_flg==1")[X], train.query("treatment_flg==1")[y])
control_model.fit(train.query("treatment_flg==0")[X], train.query("treatment_flg==0")[y])

# 예측 (확률값)
treated_prob = treated_model.predict_proba(X_test)[:,1]  # P(Y=1|T=1, X)
control_prob = control_model.predict_proba(X_test)[:,1]  # P(Y=1|T=0, X)

# CATE = 두 확률값 차이
cate = treated_prob - control_prob
print(f"Estimated ATE: {np.mean(cate) * 100:.2f}%")
```

- Y가 범주형이지만 0,1인 경우
 - 다음과 같이 Y에 대한 예측값이 확률 형태로 구해짐.
 - CATE(x)=E[Y∣T=1,X=x]−E[Y∣T=0,X=x]
 - 처치그룹의 성공확률:  E[Y∣T=1,X=x]
 - 비처치 그룹의 성공확률: E[Y∣T=0,X=x]
 - Regressor가 적합
 - Classifier를 사용하는 경우 predict_proba등을 활용

- 위의 내용을 하나의 함수로 구현: XGBTRegressor

```python
from causalml.inference.meta import XGBTRegressor

# Initialize and fit the T-Learner with XGBoost
learner_t = XGBTRegressor()
learner_t.fit(X=X_train, treatment=trmt_train, y=y_train)

# Predict the treatment effect and calculate the Average Treatment Effect (ATE)
treatment_effect = learner_t.predict(X=X_test)
ate = treatment_effect.mean()

# Output the estimated ATE as a percentage
print(f"Estimated ATE: {ate * 100:.2f}%")
```

- T러너 방식의 업리프트 모델링

```python
# T러너 방식의 업리프트 모델 학습, control_name=0은 통제군을 기준으로 학습함을 의미
t_learner_xgb = BaseTClassifier(learner=XGBClassifier(random_state=24,n_estimators=50),control_name=0) #범용 메타러너 기반 업리프트 모델 학습

t_learner_xgb.fit(X=train[X], treatment=train[T], y=train[y]) #treated, controlled  2개의 모형 학습됨
ate_t_xgb = t_learner_xgb.estimate_ate(X=test[X], treatment=test[T], y=test[y]) #ATE 추정 E[Y∣T=1]−E[Y∣T=0]
ite_t_xgb = t_learner_xgb.predict(X=test[X], treatment=test[T]) #개별 처리 효과 추정  ITE (개별 처리 효과)

```

```python
ate_t_xgb
ite_t_xgb
```

```python
result = pd.DataFrame(ite_t_xgb, columns=['treatment'])
result.shape
```

```python
#예측된 효과가 0보다 작으면 'c'(control)', 그렇지 않으면 treatment .
best_treatment = np.where((result < 0).all(axis=1), 'c', result.idxmax(axis=1))
best_treatment
```

```python
#실제 treatment가 예측한 best treatment와 일치하는 데이터 골라냄. control 데이터도 포함. AUUC/Qini 계산용 synthetic 샘플 생성.
actual_is_best = np.where(test[T].replace([0,1],['c','treatment']) == best_treatment, 1, 0) #예측과 실제가 같으면 1 아니면 0
actual_is_control = np.where(test[T].replace([0,1],['c','treatment']) == 'c', 1, 0) #실제 test의 treatment컬럼이 control이면 1 아니면 0
synthetic = (actual_is_best == 1) | (actual_is_control == 1)  #정분류하였거나, 실제 control인 경우 선택
synth = result[synthetic] #AUUC/Qini 계산용 생성
```

- (actual_is_best == 1) | (actual_is_control == 1)
 - actual_is_best == 1
 -“정분류한 샘플”선택: uplift 모델이 실제로 ‘맞춘 것들’만 성능 평가에 포함
 - AUUC/Qini는 모델이 추천한 treatment를 적용했을 때의 효과를 보고자 하며, 모델이 맞춘 샘플만 포함시켜야 실제 uplift가 얼마인지 평가

 - actual_is_control == 1
  - 실제 control 샘플도 무조건 포함: 모델이 추천을 잘못했더라도 통제군(control)은 비교 기준으로 사용하기 때문. AUUC/Qini는 treatment와 control의 uplift 차이로 계산되며, control 샘플이 빠지면 비교가 불가. 즉, “모델의 추천”이 맞든 틀리든, control은 baseline 역할.

```python
print(result[8:12])
print(test[T][8:12])
print(best_treatment[8:12])
print(actual_is_best[8:12]) #예측값이 실제 test의 T와 같은지 체크해서 1 또는 0
print(actual_is_control[8:12]) #예측값이 실제 control인지를 체크
print(synthetic[8:12])
```

```python
#AUUC 평가용 데이터 생성:
#is_treated: 실제 처리여부
#target: 실제 결과
#uplift_tree: 각 샘플의 uplift 값

#synth데이터 프레임에 컬럼 추가
auuc_metrics = (synth.assign(is_treated = 1 - actual_is_control[synthetic],
                             target = test.loc[synthetic, 'target'].values.astype(int), #test의 target값
                             uplift_tree = synth.max(axis=1))) #44720개
auuc_metrics
```

```python
 # Import necessary libraries
import numpy as np
import pandas as pd
from causalml.inference.tree import UpliftRandomForestClassifier
from causalml.metrics import plot_gain, plot_qini
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trmt_train.replace({0:"control", 1:"treatment"}, inplace=True)
```

```python
# Train Uplift Random Forest model
uplift_rf = UpliftRandomForestClassifier(control_name='control')
uplift_rf.fit(X_train.values, treatment=trmt_train.values, y=y_train.values)
```

```python
# Predict uplift using the trained model
y_pred = uplift_rf.predict(X_test)
```

- predict에 X만 넣는 이유
 - 학습(fit) 때는 treatment가 필요: 모델이 treatment 그룹과 control 그룹의 차이를 학습해야 하기 때문에, 두 그룹을 따로 모델링하거나, 그룹 차이를 분리해서 트리 분할을 설계.
 - 예측(predict):모델은 이미 “같은 X가 treatment일 때와 control일 때 결과 차이”를 학습했기 때문에 넣지 않음.
 - 즉, 하나의 X에 대해 두 확률을 내부적으로 동시에 계산해 uplift = 차이를 반환.
 - P(Y∣T=1,X)
 - P(Y∣T=0,X)
 - treatment를 넣으면 uplift가 아니라 단순한 예측값(예:P(Y∣T=treatment,X))

```python
from causalml.inference.tree import uplift_tree_string, uplift_tree_plot

# Specify a tree in the random forest (the index can be any integer from 0 to n_estimators-1)
uplift_tree = uplift_rf.uplift_forest[0]

# Print uplift tree as a string
result = uplift_tree_string(uplift_tree.fitted_uplift_tree, X_train.columns)
```
